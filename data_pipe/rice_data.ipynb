```python
overall_description = pd.read_csv(f'{log_dir}/overall_description.csv')
onlist_condition = (overall_description['de_listed_date'] == '0000-00-00')
onlist_stock_code = overall_description.loc[onlist_condition, 'order_book_id']
stock_list = onlist_stock_code.values.tolist()
```

```python
date_array = pd.read_csv(f'{log_dir}/index_kline_day/000001.XSHG.csv')
date_array = date_array.loc[(date_array['date'] >= start_time) & (date_array['date'] < end_time), 'date'].values
```

```python
def download_daily_data(stock_name, start_time, end_time, date_array, save_path = None):
    
    download_df = rqdatac.get_price(stock_name, 
                  start_date = start_time,
                  end_date = end_time,
                  frequency = '1d',
                  fields = None,
                  adjust_type = 'pre',
                  skip_suspended = True,
                  market = 'cn',
                  expect_df = True)
    try:
        fields = download_df.columns.to_list()
        download_df.reset_index(inplace = True)
        download_df.set_index('date', inplace = True)

        full_df = pd.DataFrame(columns = ['c_code', 'date'] + fields)
        full_df['date'] = date_array
        full_df.set_index('date', inplace = True)
        full_df['c_code'] = stock_name
        full_df[fields] = download_df[fields]
        if save_path is None:
            return full_df
        else:
            full_df.to_csv(f'{save_path}/{stock_name}.csv')
    except Exception as e:
        print(stock_name)
    
```

```python
save_path = f'{log_dir}/kline_day'
```

```python
for stock_name in tqdm(stock_list):
    download_daily_data(stock_name, start_time, end_time, date_array, save_path)
```

### 根据日频数据清理出一份合适的股票名单

```python
existing_stock_list = os.listdir(f'{log_dir}/kline_day')
existing_stock_list = [x[:-4] for x in existing_stock_list]
existing_stock_list.sort()
```

```python
valid_stock_list = []
null_patience = 0.05
start_time = '2015-01-01'
end_time = '2023-04-01'
```

```python
for stock_name in tqdm(existing_stock_list):
    df = pd.read_csv(f'{log_dir}/kline_day/{stock_name}.csv', index_col = 'date')
    df.loc[(df['volume'] < 1)|(df['total_turnover'] < 1), :] = np.nan
    df = df.loc[start_time:end_time, :]
    if (df.isna().sum().max()) < (null_patience * df.shape[0]):
        valid_stock_list.append(stock_name)
```

```python
len(valid_stock_list)
```

```python
valid_stock_list.sort()
```

```python
stock_index_df = pd.DataFrame(columns=range(1, len(valid_stock_list) + 1), index = date_array)
stock_index_df.index.name = 'date'
```

```python
for date in date_array:
    stock_index_df.loc[date, :] = valid_stock_list
```

```python
stock_index_df.to_csv(f'{log_dir}/stock_pool.csv')
```
