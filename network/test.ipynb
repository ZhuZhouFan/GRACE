```python
import torch
import pandas as pd
import numpy as np

from FTGCN import FTGCN_Agent
```

```python
lag_order = 16
P = 10
N = 1682 + 5
data_path = f'/data/GRACE_data/A_share/tensors/FF5/lag_{lag_order}_horizon_1'
```

```python
torch.cuda.set_device(0)
start_time = '2015-01-01'
valid_time = '2019-01-01'
end_time = '2021-01-01'
log_dir = '/data/GRACE_data/A_share/model_test'
tau = 0.0
num_workers = 3
```

```python
rel_encoding = np.load('/data/GRACE_data/A_share/factor_augmented_sector_ind_graph.npy')
```

```python
agent = FTGCN_Agent(individual_num=N,
                       feature_dim=P,
                       hidden_dim=64,
                       rel_encoding = rel_encoding, 
                       log_dir=log_dir,
                       learning_rate=1e-4,
                       seed=42, 
                       )
```

```python
agent.load_data(data_path, start_time, valid_time, end_time, num_workers)
```

```python
agent.train(tau=tau, 
            epoch=1000,
            lambda_=0.1,
            mse_loss=True)
```
